```python
import pandas as pd
import seaborn as sns
```

```python
vendas = pd.read_excel("/content/varejo.xlsx")
vendas.head()
```

```python
vendas.info()
```

```python
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```python
vendas["idcanalvenda"] = vendas["idcanalvenda"].str.replace("APP", "Aplicativo")
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```python
vendas.groupby("bandeira").idcompra.nunique()
```

```python
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```python
vendas.groupby("estado").idcompra.nunique()
```

```python
vendas["Nome_Departamento"] = vendas["Nome_Departamento"].str.replace(" ","_")
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```python
vendas.info()
```

```python
vendas[vendas["estado"].isnull()]
```

```python
vendas_sem_nulos = vendas.dropna()
vendas_sem_nulos.info()
```

```python
vendas["estado"] = vendas["estado"].fillna("MS")
```

```python
vendas["estado"].fillna("MS", inplace=True)
```

```python
vendas.info()
```

```python
media_preco = vendas["Preço"].mean()

vendas["Preço"].fillna(media_preco, inplace= True)
```

```python
media_preco
```

```python
vendas.info()
```

```python
mediana_preco = vendas["Preço"].median()
mediana_preco
```

```python
vendas.head()
```

```python
vendas[["Preço","Preço_com_frete"]].describe()
```

```python
preco_errado = vendas.query("Preço > Preço_com_frete")

vendas_correto = vendas.query("Preço < Preço_com_frete")
```

```python
vendas_correto.head()
```

```python
vendas_correto.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP'").describe()
```

```python
vendas_correto.groupby("Nome_Departamento").idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
round(vendas_correto.groupby("Nome_Departamento")["Preço_com_frete"].mean().sort_values(ascending=False).reset_index(),2)
```

```python
vendas_correto.groupby("Data").idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
vendas_correto["Mes"] = vendas_correto["Data"].dt.month
vendas_correto.head()
```

```python
vendas_correto = vendas_correto.drop(columns="mes")
vendas_correto.head()
```

```python
vendas_correto.groupby("Mes").idcompra.nunique().reset_index()
```

```python
cliente = pd.read_excel("cliente_varejo.xlsx")
cliente.head()
```

```python
cliente = cliente.astype({"renda": "float"})
cliente.info()
```

```python
vendas_cliente = vendas_correto.merge(cliente, how="left", on="cliente_Log")
vendas_cliente.head()
```

```python
agg_idcanal_renda = round(vendas_cliente.groupby("idcanalvenda")["renda"].mean().sort_values(ascending=False).reset_index(),2)
agg_idcanal_renda
```

```python
agg_idade_bandeira = round(vendas_cliente.groupby("bandeira")["idade"].mean().reset_index(),2)
agg_idade_bandeira
```

```python
import matplotlib.pyplot as plt
```

```python
fig= plt.figure(figsize=(9,7))
plt.bar(agg_idade_bandeira["bandeira"], agg_idade_bandeira["idade"], color ="black")
plt.ylabel("Média de idade")
plt.title("Média de idade por bandeira", loc="center", fontsize=10, fontweight="bold", color= "black");
```

```python
fig= plt.figure(figsize=(9,7))
plt.bar(agg_idcanal_renda["idcanalvenda"], agg_idcanal_renda["renda"], color ="black")
plt.ylabel("Média de renda")
plt.title("Média de renda por canal de venda", loc="center", fontsize=10, fontweight="bold", color= "black");
```

```python
venda_por_data = vendas_correto.groupby("Data").idcompra.nunique().reset_index()
venda_por_data
```

```python
plt.figure(figsize=(35,11))

plt.plot(venda_por_data["Data"], venda_por_data["idcompra"], color="black")
plt.ylabel("Quantidade de vendas")
plt.title("Quantidade de venda por data", loc="center", fontsize=35, fontweight="bold", color= "black");
```

```python
deph_preco = round(vendas_correto.groupby("Nome_Departamento")["Preço_com_frete"].mean().sort_values(ascending=False).reset_index(),2)
deph_preco
```

```python
plt.figure(figsize=(35,11))

plt.bar(deph_preco["Nome_Departamento"], deph_preco["Preço_com_frete"], color="black")
plt.ylabel("Preço médio")
plt.title("Preço médio por departamento", loc="center", fontsize=35, fontweight="bold", color= "black");
```
